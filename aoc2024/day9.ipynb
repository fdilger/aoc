```python
from typing import List, Tuple, Dict, Set, Any
from collections import defaultdict, deque
import heapq
import re
from copy import deepcopy

def read_input(filename: str) -> str:
    """Read input file and return contents as string."""
    with open(filename, 'r') as f:
        return f.read().strip()

def read_input_lines(filename: str) -> List[str]:
    """Read input file and return list of lines with whitespace stripped."""
    with open(filename, 'r') as f:
        return [line.strip() for line in f.readlines()]

def read_input_ints(filename: str) -> List[int]:
    """Read input file and return list of integers."""
    return [int(x) for x in read_input_lines(filename)]

def get_ints(text: str) -> List[int]:
    """Extract all integers from a string."""
    return [int(x) for x in re.findall(r'-?\d+', text)]

def manhattan_distance(p1: Tuple[int, int], p2: Tuple[int, int]) -> int:
    """Calculate Manhattan distance between two points."""
    return abs(p1[0] - p2[0]) + abs(p1[1] - p2[1])

def grid_to_dict(grid: List[List[Any]]) -> Dict[Tuple[int, int], Any]:
    """Convert 2D grid to dictionary with (row, col) coordinates as keys."""
    return {(i, j): val 
            for i, row in enumerate(grid)
            for j, val in enumerate(row)}

def get_adjacent(point: Tuple[int, int], include_diagonals: bool = False) -> List[Tuple[int, int]]:
    """Get adjacent coordinates (optionally including diagonals)."""
    x, y = point
    adjacent = [(x+1, y), (x-1, y), (x, y+1), (x, y-1)]
    if include_diagonals:
        adjacent.extend([(x+1, y+1), (x+1, y-1), (x-1, y+1), (x-1, y-1)])
    return adjacent

def bfs(start: Any, get_neighbors_fn, is_target_fn) -> Tuple[bool, Dict[Any, Any]]:
    """
    Breadth-first search returning if target found and parent dictionary for path reconstruction.
    
    Args:
        start: Starting node
        get_neighbors_fn: Function that returns neighbors for a given node
        is_target_fn: Function that returns True if node is target
    
    Returns:
        Tuple of (target_found, parent_dict)
    """
    queue = deque([start])
    visited = {start}
    parent = {}
    
    while queue:
        current = queue.popleft()
        
        if is_target_fn(current):
            return True, parent
            
        for neighbor in get_neighbors_fn(current):
            if neighbor not in visited:
                visited.add(neighbor)
                parent[neighbor] = current
                queue.append(neighbor)
                
    return False, parent

def dijkstra(start: Any, get_neighbors_with_costs_fn) -> Dict[Any, int]:
    """
    Dijkstra's algorithm for finding shortest paths.
    
    Args:
        start: Starting node
        get_neighbors_with_costs_fn: Function that returns list of (neighbor, cost) tuples
    
    Returns:
        Dictionary of distances to each node
    """
    distances = defaultdict(lambda: float('inf'))
    distances[start] = 0
    pq = [(0, start)]
    visited = set()
    
    while pq:
        current_dist, current = heapq.heappop(pq)
        
        if current in visited:
            continue
            
        visited.add(current)
        
        for neighbor, cost in get_neighbors_with_costs_fn(current):
            distance = current_dist + cost
            
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))
                
    return distances

def reconstruct_path(parent: Dict[Any, Any], target: Any) -> List[Any]:
    """Reconstruct path from parent dictionary returned by search algorithms."""
    path = []
    current = target
    while current in parent:
        path.append(current)
        current = parent[current]
    path.append(current)
    return list(reversed(path))

def rotate_grid_clockwise(grid: List[List[Any]]) -> List[List[Any]]:
    """Rotate a 2D grid 90 degrees clockwise."""
    return list(zip(*grid[::-1]))

def find_cycle_length(sequence: List[Any]) -> Tuple[int, int]:
    """
    Find the prefix length and cycle length in a sequence using Floyd's algorithm.
    Returns tuple of (prefix_length, cycle_length).
    """
    # Find a repetition
    tortoise = 1
    hare = 2
    while sequence[tortoise] != sequence[hare]:
        tortoise += 1
        hare += 2
    
    # Find start of cycle
    mu = 0
    tortoise = 0
    while sequence[tortoise] != sequence[hare]:
        tortoise += 1
        hare += 1
        mu += 1
    
    # Find cycle length
    lam = 1
    hare = tortoise + 1
    while sequence[tortoise] != sequence[hare]:
        hare += 1
        lam += 1
    
    return mu, lam
```

```python
def find_none(lst):
    for k in range(len(lst)):
        if lst[k] == None:
            return k
    return False
```

```python
def check(fs):
    i = 0
    for k in range(len(fs)):
        if fs[k] is not None:
            i += 1
            continue
        else:
            break
    for k in range(i,len(fs)):
        if fs[k] is not None:
            return True
    return False
        
    
```

```python
def checksum(fs):
    checksum = 0
    for k in range(len(fs)):
        if fs[k] is not None:
            checksum += k*fs[k]
    return checksum
```

```python
def parse_disk_map(disk_map):
    """
    Parses the disk map string into a list of blocks.
    File blocks are represented by their file ID (as strings),
    and free spaces are represented by '.'.
    
    Parameters:
        disk_map (str): The disk map string.
        
    Returns:
        List[str]: A list representing the disk blocks.
    """
    blocks = []
    file_id = 0
    is_file = True  # Start with file length
    index = 0

    while index < len(disk_map):
        length_char = disk_map[index]
        if not length_char.isdigit():
            raise ValueError(f"Invalid character '{length_char}' in disk map.")
        
        length = int(length_char)
        if is_file:
            blocks.extend([str(file_id)] * length)
            file_id += 1
        else:
            blocks.extend(['.'] * length)
        is_file = not is_file
        index += 1

    return blocks

def find_leftmost_free_block(blocks):
    """
    Finds the index of the leftmost free block in the disk.
    
    Parameters:
        blocks (List[str]): The disk blocks.
        
    Returns:
        int or None: The index of the leftmost free block, or None if no free block exists.
    """
    for i, block in enumerate(blocks):
        if block == '.':
            return i
    return None

def find_rightmost_file_block(blocks):
    """
    Finds the index of the rightmost file block in the disk.
    
    Parameters:
        blocks (List[str]): The disk blocks.
        
    Returns:
        int or None: The index of the rightmost file block, or None if no file blocks exist.
    """
    for i in reversed(range(len(blocks))):
        if blocks[i] != '.':
            return i
    return None

def compact_disk(blocks):
    """
    Compacts the disk by moving file blocks from the right end to the leftmost free space.
    
    Parameters:
        blocks (List[str]): The disk blocks.
        
    Returns:
        List[str]: The compacted disk blocks.
    """
    while True:
        left_free = find_leftmost_free_block(blocks)
        right_file = find_rightmost_file_block(blocks)

        if left_free is not None and right_file is not None and left_free < right_file:
            # Move the file block from right_file to left_free
            blocks[left_free] = blocks[right_file]
            blocks[right_file] = '.'
            # Uncomment the following line to see each move
            # print(f"Moved file '{blocks[left_free]}' from position {right_file} to {left_free}")
        else:
            # No more moves can be made
            break

    return blocks

def calculate_checksum(blocks):
    """
    Calculates the filesystem checksum by summing the product of each block's position and its file ID.
    Free spaces ('.') are skipped.
    
    Parameters:
        blocks (List[str]): The disk blocks.
        
    Returns:
        int: The computed checksum.
    """
    checksum = 0
    for position, block in enumerate(blocks):
        if block != '.':
            checksum += position * int(block)
    return checksum

def main():
    import sys

    # For demonstration, we'll use the provided example disk map
    # To use custom input, you can uncomment the input reading lines
    disk_map = read_input_lines('input9.txt')[0]

    # Uncomment the following lines to read from standard input
    # print("Enter the disk map:")
    # disk_map = sys.stdin.read().strip()

    # Parse the disk map
    try:
        blocks = parse_disk_map(disk_map)
    except ValueError as e:
        print(f"Error parsing disk map: {e}")
        return

    print("Initial Disk Blocks:")
    print(''.join(blocks))
    print(f"Total Blocks: {len(blocks)}")

    # Compact the disk
    compacted_blocks = compact_disk(blocks.copy())  # Use a copy to preserve the original for debugging
    print("\nCompacted Disk Blocks:")
    print(''.join(compacted_blocks))
    print(f"Total Blocks: {len(compacted_blocks)}")

    # Calculate the checksum
    checksum = calculate_checksum(compacted_blocks)
    print("\nFilesystem Checksum:")
    



main()
```
