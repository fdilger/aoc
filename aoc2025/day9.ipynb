```python
def polygon_area(nodes):
    n = len(nodes)
    area = 0
    for i in range(n):
        x1, y1 = nodes[i]
        x2, y2 = nodes[(i + 1) % n]
        area += x1 * y2 - x2 * y1
    return abs(area) / 2
```

```python
def check_green(x,t):
    a = x[0]
    b = x[1]
    if (a,b) in gr:
        return True
    else:
        a1 = False
        for k in range(t):
            if (a,b+k) in gr:
                a1 = True
                break
        if not a1:
            return False
        
        a2 = False
        for k in range(t):
            if (a,b-k) in gr:
                a2 = True
                break
        if not a2:
            return False
        a3 = False
        for k in range(t):
            if (a+k,b) in gr:
                a3 = True
                break
        if not a3:
            return False
        a4 = False
        for k in range(t):
            if (a-k,b) in gr:
                a4 = True
                break
        return a1 and a2 and a3 and a4
```

```python
def ccw(a, b, c):
    return (b[0]-a[0])*(c[1]-a[1]) - (b[1]-a[1])*(c[0]-a[0])

def segments_intersect(a, b, c, d):
    return ccw(a,b,c) * ccw(a,b,d) < 0 and ccw(c,d,a) * ccw(c,d,b) < 0

def point_in_polygon(point, polygon):
    x, y = point
    n = len(polygon)
    inside = False
    for i in range(n):
        x1, y1 = polygon[i]
        x2, y2 = polygon[(i + 1) % n]
        # Check if edge crosses the horizontal ray from (x, y) to the right
        if ((y1 > y) != (y2 > y)) and (x < (x2 - x1) * (y - y1) / (y2 - y1) + x1):
            inside = not inside
    return inside

def inside(a, b):
    x, y = a
    d, e = b
    if not point_in_polygon((x, e), i) or not point_in_polygon((d, y), i):
        return False
    edges = zip(i, i[1:] + [i[0]])
    for n1, n2 in edges:
        if segments_intersect(a, (x, e), n1, n2) or segments_intersect(a, (d, y), n1, n2):
            return False
        if segments_intersect(b, (x, e), n1, n2) or segments_intersect(b, (d, y), n1, n2):
            return False
    return True
```
