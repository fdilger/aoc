```python
@lru_cache(maxsize=None)
def beam(si,sj,d):
    if d == len(inp):
        return 1
    match inp[si][sj]:
        case '^':
            return beam(si+1,sj-1,d+1)+beam(si+1,sj+1,d+1)
        case '.':
            return beam(si+1,sj,d+1)
beam(o,p,0)
```
