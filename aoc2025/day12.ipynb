```python
def crds(x):
    cds = set()
    for i in range(len(x)):
        for j in range(len(x[0])):
            if x[i][j] == '#':
                cds.add((i,j))

    return cds

def get_transformations(coords):
    def rotate_90(shape):
        return [(y, 2 - x) for x, y in shape]
    def flip_horizontal(shape):
        return [(2 - x, y) for x, y in shape]

    results = []
    current = list(coords)
    for _ in range(4):
        results.append(tuple(sorted(current)))
        results.append(tuple(sorted(flip_horizontal(current))))
        current = rotate_90(current)
    return list(set(results))

shapes = [get_transformations(crds(k)) for k in i[:-1]]
def normalize(coords):
    ax, ay = coords[0]
    return tuple((x - ax, y - ay) for x, y in coords)
shapes = [[normalize(c) for c in cds ]for cds in shapes]
for s in shapes:
    print(len(s))
```

```python
from z3 import Solver, Bool, If, Sum, sat
def solve_z3(remaining, presents):
    remaining = set(remaining)
    s = Solver()
    
    # Generate all possible placements per piece type
    # placement_vars[p][k] = Bool for k-th possible placement of piece type p
    placement_vars = []
    placement_cells = []  # which cells each placement covers
    
    for p in range(len(presents)):
        p_vars = []
        p_cells = []
        for i, j in remaining:
            for c in shapes[p]:
                cds = shift(c, (i, j))
                if all(e in remaining for e in cds):
                    v = Bool(f'p{p}_{len(p_vars)}')
                    p_vars.append(v)
                    p_cells.append(frozenset(cds))
        placement_vars.append(p_vars)
        placement_cells.append(p_cells)
    
    # Constraint 1: exactly presents[p] placements chosen for each piece type
    for p in range(len(presents)):
        if presents[p] > 0:
            s.add(Sum([If(v, 1, 0) for v in placement_vars[p]]) == presents[p])
    
    # Constraint 2: each cell covered at most once
    for cell in remaining:
        covering = []
        for p in range(len(presents)):
            for k, cells in enumerate(placement_cells[p]):
                if cell in cells:
                    covering.append(placement_vars[p][k])
        if covering:
            s.add(Sum([If(v, 1, 0) for v in covering]) <= 1)
    print('finished setting up')
    return s.check() == sat
```

```python
from itertools import product
cache = set()

class Found(Exception):
    def __init__(self, solution):
        self.solution = solution

def remove_isolated(remaining):
    remaining = set(remaining)
    changed = True
    while changed:
        changed = False
        to_remove = []
        for x, y in remaining:
            if (x+1, y) not in remaining and (x-1, y) not in remaining and (x, y+1) not in remaining and (x, y-1) not in remaining:
                to_remove.append((x, y))
        if to_remove:
            changed = True
            for cell in to_remove:
                remaining.remove(cell)
    return frozenset(remaining)

def filter_components(remaining):
    remaining = set(remaining)
    components = []
    
    while remaining:
        start = next(iter(remaining))
        component = set()
        queue = [start]
        
        while queue:
            cell = queue.pop()
            if cell in remaining:
                remaining.remove(cell)
                component.add(cell)
                x, y = cell
                for nx, ny in [(x+1, y), (x-1, y), (x, y+1), (x, y-1)]:
                    if (nx, ny) in remaining:
                        queue.append((nx, ny))
        
        components.append(component)
    
    return frozenset().union(*[c for c in components if len(c) >= 5])

def normalize_remaining(remaining):
    if not remaining:
        return frozenset()
    min_x = min(x for x, y in remaining)
    min_y = min(y for x, y in remaining)
    return frozenset((x - min_x, y - min_y) for x, y in remaining)

def get_boundary(used, remaining):
    if not used:
        return remaining  # First piece - try everywhere
    
    boundary = set()
    for x, y in used:
        for nx, ny in [(x+1, y), (x-1, y), (x, y+1), (x, y-1)]:
            if (nx, ny) in remaining:
                boundary.add((nx, ny))
    return boundary



import signal

class Timeout(Exception):
    pass

def timeout_handler(signum, frame):
    raise Timeout()



def solve(remaining, used, presents):

    remaining = filter_components(remaining)
    remaining = normalize_remaining(remaining)

    state = (remaining, presents)
    
    if state in cache:
        return
    
    if all(p == 0 for p in presents):
        raise Found(1)
    
    if len(remaining) < sum(areas[l] * presents[l] for l in range(len(presents))):
        cache.add(state)
        return
    
    p = -1
    for b in range(len(presents)):
        if presents[b] > 0:
            p = b
            break
    
    k = 0
    for i, j in remaining:
        if k > 20:
            break
        for pi in range(p,len(shapes)):
            for c in shapes[pi]:
                cds = shift(c, (i, j))
                if all(e in remaining for e in cds):
                    p2 = list(presents)
                    p2[pi] -= 1
                    solve(frozenset(remaining - set(cds)), used | set(cds), tuple(p2))
        k+=1
    cache.add(state)

def solve_with_timeout(remaining, presents, timeout_sec=5):
    signal.signal(signal.SIGALRM, timeout_handler)
    signal.alarm(timeout_sec)
    try:
        cache.clear()
        result = solve(remaining, set(), presents)
        signal.alarm(0)
        return False  # No solution found (didn't raise Found)
    except Found:
        signal.alarm(0)
        return True
    except Timeout:
        return None  # Unknown - timed out



possible = 0
print('TODO: '+ str(len(trees)))
o = 0
for tree in trees:
    space = tree[0]
    presents = tuple(tree[1:])
    print(9*sum(presents))
    print(space[0]*space[1])
    if space[0]*space[1] >= 9*sum(presents):
        possible+=1
    try:
        board = frozenset(product(range(space[0]), range(space[1])))
        
    except Found:
        possible += 1
        print('found!!!__________________')
    print('DONE: ' + str(o)) 
    
print(possible)
```
