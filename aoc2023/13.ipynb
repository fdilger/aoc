```python
def checkMirrorV(j,pattern):
    ## we have to check min(j,len(pattern)-j) elements in each direction, for all the rows of the pattern now
    p = min(j,len(pattern[0])-j)
    diffs = []
    for i in range(len(pattern)):
        left = pattern[i][j:j+p]
        right = ''.join(list(reversed(pattern[i][j-p:j])))
        if left != right:
            diffs.append((left,right))
    if len(diffs) == 0 and 2*p == len(pattern[0])-1:
        return False
    if len(diffs) > 1:
        return False
    if len(diffs) == 1:
        l = diffs[0][0]
        r = diffs[0][1]
        count = 0
        for c in range(len(l)):
            if l[c]!=r[c]:
                count+=1
        if count > 1:
            return False
        if count == 1:
            ### return True only when we find a symmetry by changing a thing from before
            return True
    return False


```

```python
def checkMirrorH(i,pattern):
    p = min(i,len(pattern)-i)
    diffs = []
    for k in range(p):
        left = pattern[i+k]
        right = pattern[i-k-1]
        if pattern[i+k] != pattern[i-k-1]:
            diffs.append((left,right))    
    if len(diffs) == 0 and 2*p >= len(pattern)-1:
        return False
    if len(diffs) > 1:
        return False
    elif len(diffs) == 1:
        l = diffs[0][0]
        r = diffs[0][1]
        count = 0
        for a,b in zip(l,r):
            if a!=b:
                count+=1
        if count > 1:
            return False
        if count == 1:
            return True    
    return False
```

```python

res = 0
for pattern in patterns:
    ### colums first
    V = False
    for j in range(1,len(pattern[0])):
        if checkMirrorV(j, pattern):
            res+=j
            V = True
            break
    if not V:
        for i in range(1,len(pattern)):
            if checkMirrorH(i,pattern):
                res+=100*i
                break
print(res)
```
