```python
G = nx.DiGraph()
ir = [l for l in range(0,len(input))]
jr = [l for l in range(0,len(input[0]))]
print(ir)
print(jr)
for i in range(len(input)):
    for j in range(len(input[0])):

        ### (0,0) level connections
        if i+1 in ir:
            G.add_edge((i,j,0,0),(i+1,j,1,0),weight = input[i+1][j])
        if i-1 in ir:
            G.add_edge((i,j,0,0),(i-1,j,-1,0),weight = input[i-1][j])
        if j+1 in jr:
            G.add_edge((i,j,0,0),(i,j+1,0,1), weight = input[i][j+1])
        if j-1 in jr:
            G.add_edge((i,j,0,0),(i,j-1,0,-1), weight = input[i][j-1])
        # 1+ level connections
        for k in range(1,2):
            if i+1 in ir:
                G.add_edge((i,j,k,0),(i+1,j,k+1,0),weight = input[i+1][j])
        for k in range(2,5):
            if i+1 in ir:
                G.add_edge((i,j,k,0),(i+1,j,k+1,0),weight = input[i+1][j])
            if j+1 in jr:
                G.add_edge((i,j,k,0),(i,j+1,0,1), weight = input[i][j+1])
            if j-1 in jr:
                G.add_edge((i,j,k,0),(i,j-1,0,-1), weight = input[i][j-1])
        if j+1 in jr:
            G.add_edge((i,j,5,0),(i,j+1,0,1), weight = input[i][j+1])
        if j-1 in jr:
            G.add_edge((i,j,5,0),(i,j-1,0,-1), weight = input[i][j-1])


        # (-1+,0) level connections
        for k in range(1,2):
            if i-1 in ir:
                G.add_edge((i,j,-k,0),(i-1,j,-k-1,0),weight = input[i-1][j])
        for k in range(2,5):
            if i-1 in ir:
                G.add_edge((i,j,-k,0),(i-1,j,-k-1,0),weight = input[i-1][j])
            if j+1 in jr:
                G.add_edge((i,j,-k,0),(i,j+1,0,1), weight = input[i][j+1])
            if j-1 in jr:
                G.add_edge((i,j,-k,0),(i,j-1,0,-1), weight = input[i][j-1])
        if j+1 in jr:
            G.add_edge((i,j,-5,0),(i,j+1,0,1), weight = input[i][j+1])
        if j-1 in jr:
            G.add_edge((i,j,-5,0),(i,j-1,0,-1), weight = input[i][j-1])

        # (0,-1+) level connections
        for k in range(1,2):
            if j-1 in jr:
                G.add_edge((i,j,0,-k),(i,j-1,0,-k-1),weight = input[i][j-1])
        for k in range(2,5):
            if i-1 in ir:
                G.add_edge((i,j,0,-k),(i-1,j,-1,0),weight = input[i-1][j])
            if i+1 in ir:
                G.add_edge((i,j,0,-k),(i+1,j,1,0), weight = input[i+1][j])
            if j-1 in jr:
                G.add_edge((i,j,0,-k),(i,j-1,0,-k-1), weight = input[i][j-1])
        if i+1 in ir:
            G.add_edge((i,j,0,-5),(i+1,j,1,0), weight = input[i+1][j])
        if i-1 in ir:
            G.add_edge((i,j,0,-5),(i-1,j,-1,0), weight = input[i-1][j])

        # (0,1+) level connections
        for k in range(1,2):
            if j+1 in jr:
                G.add_edge((i,j,0,k),(i,j+1,0,k+1),weight = input[i][j+1])
        for k in range(2,5):
            if i-1 in ir:
                G.add_edge((i,j,0,k),(i-1,j,-1,0),weight = input[i-1][j])
            if i+1 in ir:
                G.add_edge((i,j,0,k),(i+1,j,1,0), weight = input[i+1][j])
            if j+1 in jr:
                G.add_edge((i,j,0,k),(i,j+1,0,k+1), weight = input[i][j+1])
        if i+1 in ir:
            G.add_edge((i,j,0,5),(i+1,j,1,0), weight = input[i+1][j])
        if i-1 in ir:
            G.add_edge((i,j,0,5),(i-1,j,-1,0), weight = input[i-1][j])
        
```
