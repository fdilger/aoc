```python
input = input.split('\n')
input = [re.findall(r'.|#|>|<|v|^', line) for line in input]
```

```python
ir = list(range(len(input)))
jr = list(range(len(input[0])))
print(ir)
print(jr)
```

```python
grid = input.copy()
```

```python
import networkx as nx
import heapq
G = nx.Graph()
```

```python

for i in ir:
    for j in jr:
        if grid[i][j] == '#':
            continue
        if i+1 in ir and grid[i+1][j] != '#':
            G.add_edge((i,j),(i+1,j),weight = 1)
        if i-1 in ir and grid[i-1][j] != '#':
            G.add_edge((i,j),(i-1,j), weight = 1)
        if j+1 in jr and grid[i][j+1] != '#':
            G.add_edge((i,j),(i,j+1),weight = 1)
        if j-1 in ir and grid[i][j-1] != '#':
            G.add_edge((i,j),(i,j-1),weight = 1)
```

```python
interesting = []
for node in G.nodes():
    count = 0
    for (i,j) in G.neighbors(node):
        if grid[i][j] in ['>','<','v','^']:
            count+=1
    if count > 2:
        interesting.append(node)
len(interesting)
```

```python
interesting.append((0,1))
interesting.append((140,139))
```

```python
H = nx.Graph()
```

```python
visited = set()
```

```python
def explore(G,s,path,k):
    if s in interesting:
        H.add_edge(path[0],s,weight = k)
        return
    else:
        for n in G.neighbors(s):
            if n not in path:
                explore(G,n,path+[s],k+1)
```

```python
for i in interesting:
    for n in G.neighbors(i):
        explore(G,n,[i],1)
```
