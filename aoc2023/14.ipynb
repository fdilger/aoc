```python
def moveN(i,j,board):
    while(1):
        if i-1>=0:
            if board[i-1][j] == '.':
                board[i-1][j] = 'O'
                board[i][j] = '.'
                i=i-1
                continue
            else:
                break
        else:
            break    
    return len(board)-i
```

```python
def moveE(i,j,board):
    while(1):
        if j+1<len(board[0]):
            if board[i][j+1] == '.':
                board[i][j+1] = 'O'
                board[i][j] = '.'
                j=j+1
                continue
            else:
                break
        else:
            break    
```

```python
def printBoard(board):
    time.sleep(0.3)
    ret = ''
    for i in range(len(board)):
        s = '     '
        for j in range(len(board[0])):
            if board[i][j] == 'O':
                s+= ' ' + ('\033[35m'+board[i][j] +'\033[0m')
            if board[i][j] == '.':
                s +=' '+('\033[37m'+board[i][j] +'\033[0m')
            if board[i][j] == '#':
                s+=' '+ ('\033[34m'+ board[i][j] +'\033[0m')
        s+='\n'
        ret+=s
    print('\n')
    print(ret)
    print('\n')
```
