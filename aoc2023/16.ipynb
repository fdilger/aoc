```python
import time
def printBoard(input):
    time.sleep(0.15)
    text = ''
    for i in range(len(input)):
        line = '                     '
        for j in range(len(input[0])):
            if input[i][j] == '#':
                line+= ('\033[93m'+input[i][j] +'\033[0m')
            elif input[i][j] == '.':
                line+= ('\033[37m'+input[i][j] +'\033[0m')
            elif input[i][j] == '/':
                line+= ('\033[94m'+input[i][j] +'\033[0m')
            elif input[i][j] == '\\':
                line+= ('\033[96m'+input[i][j] +'\033[0m')
            elif input[i][j] == '-':
                line+= ('\033[97m'+input[i][j] +'\033[0m')
            elif input[i][j] == '|':
                line+= ('\033[95m'+input[i][j] +'\033[0m')
            else:
                line+= ('\033[96m'+ input[i][j] +'\033[0m')
        line += '\n'
        text+=line
    with open("log.txt", "a") as log_file:
        print('\n', file = log_file)
        print('\n', file = log_file)
        print(text, file=log_file)
        print('\n', file = log_file)
        print('\n', file = log_file)
```
