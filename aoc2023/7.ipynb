```python
def f(h):    
    for i in range(len(h)):
            if h[i] == 'A':
                h[i] = 'Z'
            if h[i] == 'K':
                h[i] = 'Y'
            if h[i] == 'Q':
                h[i] = 'X'
            if h[i] == 'J':
                h[i] = '1'
            if h[i] == 'T':
                h[i] = 'V'
    return h
```

```python
ahands = []
for h in hands:
    dict = {}
    j = 0
    for c in h:
        if c == 'J':
            j = 1
        if c in dict:
            dict[c]+=1
        else:
            dict[c] = 1

    if len(dict) == 1:
        h = f(h)
        ahands.append((h,7))
    if len(dict) == 2:
        if dict[h[0]] == 4 or dict[h[0]] == 1:
            h = f(h)
            ahands.append((h,6+j))
        else:
            h = f(h)
            ahands.append((h,5+j*2))
    if len(dict) == 3:
        if dict[h[0]] == 3 or dict[h[1]] == 3 or dict[h[2]] == 3:
            h = f(h)
            ahands.append((h,4+2*j))
        else:
            h = f(h)
            if j: 
                if dict['J'] == 1:
                    ahands.append((h,5))
                if dict['J'] == 2:
                    ahands.append((h,6))
            else:
                ahands.append((h,3))
    if len(dict) ==4:
        h = f(h)
        if j:
            if dict['J'] == 1:
                ahands.append((h,4))
            if dict['J'] == 2:
                ahands.append((h,4))
        else:
            ahands.append((h,2))
    if len(dict) ==5:
        h = f(h)
        ahands.append((h,1+j))
                
```
